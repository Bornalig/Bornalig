```python
#Possible solution to question 4:
#Inspect first 5 rows:
data.head(5)
#Inspect last 5 rows:
data.tail(5)
#Drop NA's:
data.dropna(inplace = True)

#Replace \N in Actress and Director:

data['Actress'] = data['Actress'].replace('\\N', 'Unknown')
data['Director'] = data['Director'].replace('\\N', 'Unknown')
```

```python
#Question #5 Partial Solution:
#Filter the data:

data_1980_2000 = data[(data['Year'] >= 1980) & (data['Year'] <= 2000)]
data_1980_2000
```

```python
#Partial Solution to Question #5:
#Calculate average length of movie by year and award status:

data_1980_2000_awards= data_1980_2000.groupby(['Year', 'Awards']).Length.mean().reset_index()

data_1980_2000_awards
```

```python
#Partial Solution to #5:
#Plot the trends:

import plotly.express as px
fig = px.line(data_1980_2000_awards, x = 'Year', y = 'Length', color = 'Awards', title = 'Avg. Movie Length (Minutes) by Year (1980 - 2000)')
fig.show()
```
